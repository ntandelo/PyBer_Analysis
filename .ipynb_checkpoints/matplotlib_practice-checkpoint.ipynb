```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with pyplot (MATLAB method)
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() (quick way to plot using object-oriented method)
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and use plt.show(), BUT THIS CAN CAUSE BUGS IF USED MORE
#THAN ONCE
plt.plot(x_axis, y_axis)
#plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="^", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#scatter plot!
plt.plot(x_axis, y_axis, 'o')
```

```python
#also scatter!
plt.scatter(x_axis, y_axis)
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.scatter(x_axis, y_axis, color="red", label="Chicago")
plt.title("PyBer Fare by Month")
plt.legend()
plt.gca().invert_yaxis()
```

```python
#make a bubble plot, s scales the bubbles
plt.scatter(x_axis, y_axis, s=y_axis) #default scaling
```

```python
#make bubbles bigger than default (above)
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#list comprehension to re-size points instead of a for loop
plt.scatter(x_axis, y_axis, color="blue", label="Boston", s = [i * 3 for i in y_axis])
```

```python
#object-oriented approach
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
#bubble plot object-oriented approach
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, alpha = 0.3, edgecolors = "black", s=[i * 5 for i in y_axis], color = "blue", label = "Boston")
plt.title("PyBer Fare by Month")
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.legend()
```

```python
#MATLAB style
plt.pie(y_axis, labels=x_axis)

plt.show()
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%'
                   )

plt.show()
```

```python
#pie charts using object-oriented approach

explode_values = (0, 0, 0, 0.2, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots()
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

ax.pie(y_axis,labels=x_axis, explode=explode_values, autopct='%.1f%%', colors = colors, radius = 2.5, counterclock = False, startangle = 100, shadow = True)

plt.show()
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
#same chart using object-oriented approach
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
# plt.show()
```

```python
#bar usng MATLAB
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
#bar using object-oriented
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
import numpy as np
np.arange(0, 51, step=5.0)
```

```python
#bar with numpy, objet-oriented
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
